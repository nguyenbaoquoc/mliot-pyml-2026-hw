## 5. Task 4 — Outlier + **quyết định xử lý**

Không chỉ phát hiện, phải **quyết định** làm gì với chúng.

```python
def iqr_outliers(s):
    q1, q3 = s.quantile(.25), s.quantile(.75)
    iqr = q3 - q1
    low, high = q1 - 1.5*iqr, q3 + 1.5*iqr
    return ((s < low) | (s > high)).sum(), (low, high)

for col in ["Age", "Fare"]:
    n_out, (lo, hi) = iqr_outliers(df[col].dropna())
    z = np.abs(stats.zscore(df[col].dropna()))
    print(f"{col:5s}: IQR outliers = {n_out:3d} (ngoài [{lo:.1f}, {hi:.1f}]) | Z-score(>3) = {(z>3).sum()}")
```

**Quyết định:** `Fare` có nhiều outlier theo IQR, nhưng đó là **vé hạng nhất có thật** (giá trị hợp lệ), không phải lỗi nhập liệu. → **Giữ lại** và dùng **RobustScaler** (bền vững với outlier) thay vì xóa. `Age` outlier không đáng kể.

## 7. Task 6 — Chia tập TRƯỚC (chống data leakage)

> **Nguyên tắc vàng:** mọi phép "học tham số" từ dữ liệu (median để điền, mode, min/max/IQR để scale, danh mục để encode) **chỉ được học từ tập train**. Vì vậy phải **chia tập trước**, rồi mới `fit` bộ tiền xử lý trên train và `transform` cho validation/test.

```python
X = df.drop(columns="Survived")
y = df["Survived"]

# 70 / 15 / 15, stratify để giữ tỷ lệ hai lớp
X_tmp, X_test, y_tmp, y_test = train_test_split(X, y, test_size=0.15, random_state=42, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_tmp, y_tmp, test_size=0.1765, random_state=42, stratify=y_tmp)
print("Train:", X_train.shape, "| Val:", X_val.shape, "| Test:", X_test.shape)
for name, yy in [("train", y_train), ("val", y_val), ("test", y_test)]:
    print(f"  tỷ lệ Survived ({name}): {yy.mean():.3f}")
```

## 8. Task 7 — Pipeline tiền xử lý (impute → encode → scale), fit CHỈ trên train

Dùng `ColumnTransformer` gộp mọi bước, `fit` một lần trên `X_train`:

- **Số** (`Age`, `SibSp`, `Parch`, `Fare`): `SimpleImputer(median)` → `RobustScaler`.
- **Phân loại** (`Sex`, `Embarked`): `SimpleImputer(mode)` → `OneHotEncoder(drop="first")`.
- **Thứ tự** (`Pclass`): giữ nguyên (ordinal 1 < 2 < 3).

> **Vì sao cách này chống leakage:** median/mode/tham số scale/danh mục encode đều được học **từ train**, rồi áp cho val & test — thông tin của val/test không "rò" ngược vào train.

```python
num_cols = ["Age", "SibSp", "Parch", "Fare"]
cat_cols = ["Sex", "Embarked"]
ord_cols = ["Pclass"]

num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                     ("scaler",  RobustScaler())])
cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                     ("onehot",  OneHotEncoder(drop="first", handle_unknown="ignore"))])

preprocess = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols),
    ("ord", "passthrough", ord_cols),
])

preprocess.fit(X_train)                       # fit CHỈ trên train
Xtr = preprocess.transform(X_train)
Xva = preprocess.transform(X_val)
Xte = preprocess.transform(X_test)
print("Sau tiền xử lý — train:", Xtr.shape, "| val:", Xva.shape, "| test:", Xte.shape)
print("Tên cột sau biến đổi:", list(preprocess.get_feature_names_out()))
```

> **Ghi chú dummy trap:** `OneHotEncoder(drop="first")` bỏ 1 cột cho mỗi biến (ví dụ `Sex` có 2 giá trị chỉ tạo 1 cột `Sex_male`) để tránh **đa cộng tuyến** — khác với minh họa "full one-hot" (Red=100...) trên slide vốn dùng cho trực giác.

## Hàm tính các giá trị metrics

```python
def caculate_metrics_classification(y_test, y_pred):
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    cm = confusion_matrix(y_test,y_pred)

    metrics = {
        "Accuracy" : round(accuracy,4),
        "Precision" : round(precision,4),
        "F1" : round(f1,4),
        "Recall" : round(recall,4)
    }

    for key, value in metrics.items():
        print(f"{key} : {value}")
    print()
    cm_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
    print(cm_df)
    return metrics, cm
```

## Train mô hình `logistic_regression` và ` linear_regression1`

```python
# Logisic_regression
Lor_regression = LogisticRegression(random_state = 42)
Lor_regression.fit(Xtr,y_train)
y_pred_lor = Lor_regression.predict(Xte)

# Linear_regression
lr_model = LinearRegression()
lr_model.fit(Xtr,y_train)
y_pred_lr = lr_model.predict(Xte)
y_pred_lr_cl = (y_pred_lr >=0.5).astype(int) 
```

## Tính toán metrics

```python
print("Metics của Logistic_regression:")
me, cm = caculate_metrics_classification(y_test,y_pred_lor)
print()
print("Metrics của Linear_regression:")
me_lr, cm_lr = caculate_metrics_classification(y_test,y_pred_lr_cl)
```
